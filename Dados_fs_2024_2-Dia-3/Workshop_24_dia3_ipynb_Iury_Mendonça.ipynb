```python
# calcule quais os valores mínimos, máximos e médios de pontuação de cada matéria
# (matemática, leitura e escrita), deixe claro cada valor comentado
# em markdown

valor_minimo = df['math score'].min()
valor_maximo = df['math score'].max()
media = df['math score'].mean()

print('valor minimo é: ' ,valor_minimo)
print('valor maximo é: ' ,valor_maximo)
print('media é: ', media)
```

valor minimo é:  0
valor maximo é:  100

```python
# Agrupe o banco de dados a partir da raça/etnia dos alunos,
# e plote um gráfico com a nota média em Matemática
# para cada grupo disponível;
df2=df.groupby('race/ethnicity')[['math score']].mean()
df2
```

```python

df2.plot(kind='bar')
```

```python
# Agrupe o banco de dados a partir do Nível de Escolaridade dos
# pais, e plote um gráfico com a nota média em Escrita para cada
# grupo disponível;

df3=df.groupby('parental level of education').count()
df3
```

```python
df3=df.groupby('race/ethnicity')[['writing score']].mean()
df3
```

```python
df3.plot(kind ='bar', subplots=True)
```

```python
# Agrupe o banco de dados a partir da coluna referente à participação em um curso preparatório
# para os testes, e plote um gráfico com a nota média em Leitura para cada grupo disponível.

# primeiro fazendo a relação entre etnia e preparação pra teste
df4=df.groupby('race/ethnicity')[['test preparation course']].count()
```

```python
df4.plot(kind='bar', subplots=True)
```

```python
#relação entre preparação e score de leitura
df5=df.groupby('test preparation course')[['reading score']].mean()
```

```python
df5.plot(kind='bar', subplots=True)
```
